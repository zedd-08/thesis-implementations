```python
results_file = 'survey_results.json'
questss_file = 'survey_quests.json'
```

```python
import json
results = json.load(open(results_file))
quests = json.load(open(questss_file))
```

```python
game_id_mappings = {
	'TESO': "The Elder Scrolls: Oblivion",
	'TESS': "The Elder Scrolls: Skyrim",
	'BG1': "Baldur's Gate 1",
	'BG2': "Baldur's Gate 2",
	'MC': "Minecraft",
	'TL2': "Torchlight II"
}
```

```python
for k, v in results.items():
	num = int(k[1]) - 1
	if k =='Q10':
		num = 9

	quest = quests[num]
	plots = quest['no_kg_input'].split('\n')[1:-2]
 
	no_kg = quest['no_kg'].split('\n')
	text_kg = quest['text_kg'].split('\n')
	actual = quest['actual'].split('\n')

	print('\\begin{itemize}')
	print('\t\\item[\\textbf{Game:}] \\textbf{' + game_id_mappings[quest['game']] + '}')
	print('\t\\item[\\textbf{Description:}] \\textbf{' + quest['quest_giver'] + ':} "' + '\n\n\t'.join(quest['desc'].split('\n')) + '"\n')
	print('\t\\item[\\textbf{Plot points:}]')
	print('\t\\begin{itemize}')
	for plot in plots:
		print('\t\t\\item ' + plot)
	print('\t\\end{itemize}\n')

	print('\t\\item[\\textbf{KG2Text:}]')
	print('\t\\begin{itemize}')
	for kkg in quest['text_kg_input'].split('\n\n\n### Plots')[0].split('\n')[1:]:
		print('\t\t\\item ' + kkg)
	print('\t\\end{itemize}\n')
 
	print('\t\\item[\\textbf{No KG output:}]\n')
	print('\t\\textbf{Title:} ' + no_kg[0][7:] + '\n')
	print('\t\\textbf{Description:} ' + no_kg[1][11:] + '\n')
	print('\t\\textbf{Tasks:}')
	print('\t\\begin{itemize}')
	for task in no_kg[3:]:
		print('\t\t\\item ' + task)
	print('\t\\end{itemize}\n')
 
	print('\t\\item[\\textbf{TextKG output:}]\n')
	print('\t\\textbf{Title:} ' + text_kg[0][7:] + '\n')
	print('\t\\textbf{Description:} ' + text_kg[1][11:] + '\n')
	print('\t\\textbf{Tasks:}')
	print('\t\\begin{itemize}')
	for task in text_kg[3:]:
		print('\t\t\\item ' + task)
	print('\t\\end{itemize}\n')
 
	print('\t\\item[\\textbf{Actual output:}]\n')
	print('\t\\textbf{Title:} ' + actual[0][7:] + '\n')
	print('\t\\textbf{Description:} ' + actual[1][11:] + '\n')
	print('\t\\textbf{Tasks:}')
	print('\t\\begin{itemize}')
	for task in actual[3:]:
		print('\t\t\\item ' + task)
	print('\t\\end{itemize}\n')
	print('\\end{itemize}')
	print('\\newpage')
	print()
```
