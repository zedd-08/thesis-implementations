```python
import pandas as pd
import numpy as np
```

```python
def load_dataset_excel(path):
    df = pd.read_excel(path)
    print(df.info())
    return df
```

```python
def load_dataset_json(path):
    df = pd.read_json(path)
    print(df.info())
    return df
```

```python
def get_avg_num_words(df_column):
    avg = df_column.apply(lambda x: np.mean([len(x.split(' '))]))
    num_q = len(list(avg))
    full_avg = np.mean(list(avg))
    print(f'Average number of words in description: {full_avg}')
    return {'avg': full_avg, 'NumOfQuests': num_q}
```

```python
averages = {}
```

# Dataset gathered and compiled by Sander Koomen

```python
averages['Sander'] = {}
```

## Elder Scrolls Online dataset

```python
elderscrolls_path = 'Sander/ElderScrollsOnlineDataset.xlsx' 
df = load_dataset_excel(elderscrolls_path)
```

```python
df = df.drop(['level', 'faction', 'repeatable'], axis=1)
df.head()
```

#### Avg num of words in all quests

```python
averages['Sander']['ElderScrollsOnline_all'] = get_avg_num_words(df.description)
```

#### Avg num of words in `Main` quests

```python
main_q = df.query('quest_type == "Main"')
main_q.head()
```

```python
averages['Sander']['ElderScrollsOnline_main'] = get_avg_num_words(main_q.description)
```
